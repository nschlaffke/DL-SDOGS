```python
from matplotlib import pyplot as plt
%matplotlib inline
prefix = 'aug'
```

```python
def is_loss(name):
    return name.find('loss') != -1
    
names = ['val_loss', 'val_acc', 'acc', 'loss']
titles = ['Test set loss', 'Test set accuracy', 'Training set accuracy', 'Training set loss']
data = dict()
for name, title in zip(names, titles):   
    with open(name + '.txt') as f:
        content = f.readlines()
    data[name] = [float(x) if is_loss(name) else float(x) * 100 for x in content]
```

```python
plt.title('Loss function value in the number of epochs')
plt.xlabel('number of epochs')
plt.ylabel('loss function value')
plt.plot(data['val_loss'], label='test set loss')
plt.plot(data['loss'], label='training set loss')
plt.legend(loc='best')
plt.savefig(prefix + '_' + 'loss.pdf')
```

```python
plt.title('Accuracy in the number of epochs')
plt.xlabel('number of epochs')
plt.ylabel('accuracy value')
plt.plot(data['val_acc'], label='test set accuracy')
plt.plot(data['acc'], label='training set accuracy')
plt.legend(loc='best')
plt.savefig(prefix + '_' + 'acc.pdf')
```

```python
for key in data.keys():
    print(key, min(data[key]) if is_loss(key) else max(data[key]) )
```
